## 2. Instalacja Pythona (Anaconda / Miniconda)
Zamiast instalować samego Pythona, użyjemy **Anaconda** lub lżejszej wersji **Miniconda**. To narzędzia, które pozwalają tworzyć „izolowane pokoje” (środowiska) dla różnych projektów, dzięki czemu biblioteki nie będą się ze sobą kłócić.

### Krok 1: Pobieranie
- **Windows/Mac:** Pobierz **Miniconda** (lekka wersja) stąd: [https://docs.anaconda.com/miniconda/](https://docs.anaconda.com/miniconda/)
- Zainstaluj, klikając „Next” (na Windowsie zaznacz opcję *Add to PATH*, jeśli instalator o to zapyta, choć zazwyczaj zalecają używanie *Anaconda Prompt*).

### Krok 2: Tworzenie środowiska (Otwórz Warp lub Terminal)
Wpisz poniższą komendę, aby stworzyć środowisko o nazwie `base_ai` z Pythonem w wersji 3.12:
```bash
conda create -n base_ai python=3.12 -y
```

Następnie aktywuj to środowisko:
```bash
conda activate base_ai
```

## 3. Instalacja potrzebnych bibliotek
Kiedy Twoje środowisko jest aktywne (`(base_ai)` powinno widnieć przed nazwą użytkownika w terminalu), zainstaluj wszystkie narzędzia potrzebne do naszych laboratoriów:

```bash
pip install numpy pandas matplotlib scikit-learn notebook ipykernel
```

Aby ten notebook widział Twoje nowe środowisko wewnątrz Jupyter, wpisz jeszcze to:
```bash
python -m ipykernel install --user --name base_ai --display-name "Python (base_ai)"
```

## 4. Co to jest Jupyter Notebook?
To, co teraz widzisz, to właśnie Jupyter Notebook. To dokument, który łączy w sobie **tekst** (Markdown) oraz **kod** (Python), który można uruchamiać fragmentami.

### Podstawy obsługi:
1. **Komórki (Cells):** Dokument składa się z bloków. Blok tekstowy to *Markdown*, a blok z kodem to *Code*.
2. **Uruchamianie:** Kliknij w komórkę z kodem i naciśnij `Shift + Enter`. Kod zostanie wykonany, a wynik pojawi się poniżej.
3. **Menu:** Na górze możesz dodawać nowe komórki (`+`) lub zmieniać ich typ.

## 5. Test Środowiska (Uruchom mnie!)
Kliknij w poniższą komórkę i naciśnij `Shift + Enter`. Jeśli zobaczysz wykres i napis „Wszystko działa!”, Twoje środowisko jest gotowe!

````python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import pandas as pd

# Prosty test rysowania
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, color='blue', label='Fala Sinus')
plt.title("Test środowiska AI")
plt.legend()
plt.grid(True)
plt.show()

print(f"Wersja NumPy: {np.__version__}")
print(f"Wersja Scikit-learn: {sklearn.__version__}")
print("\nGRATULACJE! Wszystko działa poprawnie.")
````
